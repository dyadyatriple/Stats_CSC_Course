```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis,skew
import matplotlib.pyplot as plt
import pylab
import seaborn as sns
```

```python
df=pd.read_csv('/Users/dddtriple/Downloads/13_6',delimiter='\t',header=None,names=['type','date','Buyers_6','Buyers_13','supermarket'])
```

```python
df.head()
```

type - type of action, 
date - date, 
Buyers_6 - quantity of buyers on the sixth day of month, 
Buyers_13 - quantity of buyers on the thirteenth day of month, 
supermarket - name of supermarket

```python
#task 1: find the sample size
print('Sample size equals to {}'.format(df.size))
```

```python
#task 2: find the mean,median and variance of the Buyers_13
print('sample mean={},sample median={} and sample variance={}'.format(
    np.mean(df['Buyers_13']),
    np.median(df['Buyers_13']),
    round(np.var(df['Buyers_13']),2)))
```

```python
#task 3: find the mean, kurtosis and skew of Byers_13-Byers_6
differ=df['Buyers_13']-df['Buyers_6']
print("sample mean={},sample kurtosis={} and sample skew={}".format(
    round(np.mean(differ),2),
    round(kurtosis(differ),2),
    round(skew(differ),2)))
```

```python
plt.figure(figsize=(25,15))

pylab.subplot(2,3,1)
pylab.hist(df['Buyers_6'])
pylab.title('Buyer 6')

pylab.subplot(2,3,2)
pylab.hist(df['Buyers_13'])
pylab.title('Buyer 13')

pylab.subplot(2,3,3)
pylab.hist(differ)
pylab.title('Difference')

pylab.subplot(2,3,4)
pylab.boxplot(df['Buyers_6'])

pylab.subplot(2,3,5)
pylab.boxplot(df['Buyers_13'])

pylab.subplot(2,3,6)
pylab.boxplot(differ)



plt.show()
```

```python
#part 2 college
df=pd.read_csv('/Users/dddtriple/Downloads/colleges.txt',delimiter='\t')
```

```python
df.head()
```

```python
#task 1:get the unique values of school type
df['School_Type'].unique()
```

```python
#task 2:analize the boxplots and answer the questions
plt.figure(figsize=(17,11))
for i, j in enumerate(['SAT', 'Acceptance', '$/Student', 'Top 10%', '%PhD', 'Grad%']):
    plt.subplot(2,3,i+1)
    sns.boxplot(x='School_Type', y=j, data=df)
```

```python
#task 3: analize the hists and answer the questions
uni=df[df['School_Type']=='Univ']
art=df[df['School_Type']=='Lib Arts']
plt.figure(figsize=(17,12))
for i, dat in enumerate(df.drop(labels=['School','School_Type'],axis=1).columns):
    plt.subplot(2,3,i+1)
    sns.distplot(uni[dat],label='u',color='green')
    sns.distplot(art[dat],label='a')
    plt.legend()
```

```python
#task 4: get mean of SAT at art schools,Q3 of acceptance of univ., the univ which pays the most
mean_art=art['SAT'].mean()
q3_acc=np.quantile(a=uni['Acceptance'],q=0.75)
maxi=df[df['$/Student']==df['$/Student'].max()].loc[:,'School'].values[0]
print('{}, {}, {}'.format(mean_art,q3_acc,maxi))
```
